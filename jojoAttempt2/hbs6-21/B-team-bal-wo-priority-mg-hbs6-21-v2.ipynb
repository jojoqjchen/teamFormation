## Build people table pp

```python
# use people matrix pp
# p is range from 0 to length of pp, ie 1 to N-1
# goal of teams of size m
# let N/m = number of team maximum, assume variable name S

#pp = pd.DataFrame(np.random.randint(0,100,size=(40, 4)), columns=list('ABCD'))
#p = range(len(pp))

pp = pd.read_csv('pp-hbs-6-21.final.csv')# people, pp-mg-sum21.csv
p = range(len(pp))
pp
```

```python
# set up for matching and build ppreport, a report table
pc=list(p)
ppcol = list(pp.columns)
#ppc = pp[['Level','Gender','Teamwork','Major']].copy() #deep copy,pp is unchanged

### NOTE: ppc would take into account only the columns that are not "ignored" by the user! 
ppc = pp[[ppcol[5],ppcol[6],ppcol[7],ppcol[8],ppcol[9],ppcol[10],ppcol[11],ppcol[12],ppcol[13],ppcol[14],ppcol[15],ppcol[16],ppcol[17],ppcol[18],ppcol[19],ppcol[20],ppcol[21],ppcol[22],ppcol[23]]].copy() #deep copy,pp is unchanged
#m = 5
N = len(pp)

S = 12 # S is number of teams to be created, approx N/m

### NOTE: ppreport is just a deep copy of pp?
ppreport = pp[[ppcol[1],ppcol[2],ppcol[3], ppcol[4], ppcol[5], ppcol[6],ppcol[7],ppcol[8],ppcol[9],ppcol[10],ppcol[11],ppcol[12],ppcol[13],ppcol[14],ppcol[15],ppcol[16],ppcol[17],ppcol[18],ppcol[19],ppcol[20],ppcol[21],ppcol[22],ppcol[23],ppcol[24]]].copy() #deep copy,pp is unchanged
ppreport['Team']='TBA'
ppreport['Similarity']=np.nan
ppreport

#prjmat =[] #intended list list of project team
prjmat =[]
for s in range(S): prjmat.append([])
print('prjmat:',prjmat) 
print(pc)
# print(ppreport.head())
# print(ppc.head())
ppreport.head()
```

## Assign People to Teams Iteratively

```python

for s in range(S):
    prjmat[s].append(s)
    pc.remove(s)
print(prjmat,pc)   

#add element to team i starting with 0 to S-1, and repeat
i=0

while len(pc)!=0:
    
    mcor = 1000000.0 # a large number
    #print(mcor)
    minj = -1
    
    for j in pc:
        prjcopy = prjmat[i].copy()
        prjcopy.append(j)
        tmp = ppc.iloc[prjcopy]
        ccor=tmp.transpose().corr().sum().sum() #current correlation mattrix sum
        #print(ccor)
        if ccor < mcor:
            mcor = ccor
            minj = j
            #print('remove ',j, mcor)
    
    #after loop of all pc
    if minj != -1: 
        pc.remove(minj)   # then add to prjmat[i]
        prjmat[i].append(minj)
        
    i=i+1
    if i==S: i=0
    
prjmat
```

```python
#test that no one is unassigned
pc
```

# Add team number to ppreport

```python
k=0
for i in range(S):
    for j in range(len(prjmat[i])):
        #print(prjmat[i][j])
        ppreport.at[prjmat[i][j], 'Team'] = i #prjmat[i][j]
        k=k+1
ppreport      
```
